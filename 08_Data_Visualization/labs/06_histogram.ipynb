## Data Wrangling

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# read from excel and clean up index
df = pd.read_excel('canada_clean.xlsx')
df.rename(columns={'Unnamed: 0':'Country'}, inplace=True)
df.set_index('Country', inplace=True)

years = list(map(str, range(1980, 2014)))

# top 5 origin countries for immigration to canada
df.sort_values(['Total'], ascending=False, inplace=True)

# set ggplot style
mpl.style.use('ggplot')
```

# Histogram

```python
# np.histogram returns 2 values
count, bin_edges = np.histogram(df['2013'])

print('Count:', count) # frequency count
print('Bin edges:', bin_edges) # bin ranges, default = 10 bins
```

```python
df['2013'].plot(kind='hist', figsize=(8, 5))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

## Setting Marks on xAxis (xticks)

```python
# 'bin_edges' is a list of bin intervals
count, bin_edges = np.histogram(df['2013'])

df['2013'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Immigration from 195 countries in 2013') # add a title to the histogram
plt.ylabel('Number of Countries') # add y-label
plt.xlabel('Number of Immigrants') # add x-label

plt.show()
```

## Setting Bins manually

```python
df['2013'].plot(kind='hist',
                figsize=(8, 5),
                bins=list(range(0,35001, 5000)))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
df_t = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
count, bin_edges = np.histogram(df_t, 15)

# un-stacked histogram
df_t.plot(kind ='hist', 
          figsize=(10, 6),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )


plt.show()
```

### Compare this to a Line Chart

```python
df_t.plot(kind='line', figsize=(10, 6))
plt.title('Line Chart of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```
